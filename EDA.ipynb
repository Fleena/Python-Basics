## Outlier Analysis

```python
df.boxplot();
```

```python
for i in numCol:
    plt.figure()
    sns.boxplot(x=df[i])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum())
```

```python
outNum = []
for i in numCol:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3 - q1
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr
    outNum.append([i, q1, q3, iqr, lower_limit, upper_limit])

    df[i] = np.where(df[i] > upper_limit, upper_limit,
                  np.where(df[i] < lower_limit, lower_limit,
                      df[i]))
```

```python
pd.DataFrame(outNum, columns=["variableName","Q1","Q3","IQR","lower_limit","upper_limit"])
```
