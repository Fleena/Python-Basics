### Load Sales data

```python
salesData = pd.read_excel("dataFiles/Sales_Data.xlsx")
salesData
```

```python
salesData.shape
```

```python
salesData.columns
```

```python
salesData.info()
```

```python
numCol = []
catCol = []

for col in salesData.columns:
    if salesData[col].dtype == 'O':
        catCol.append(col)
    else:
        numCol.append(col)
```

```python
print("List of categorical columns:", catCol)
print("List of numerical columns:", numCol)
```

```python
salesData.describe()
```
