```python
# 1. Write a python code to open a file for appending data.

fileobj = open("Sample_txt.txt","a")
fileobj.write("\n Ola ! Now there is a new appended content")
fileobj.close() 
```

```python
fileobj = open("Sample_txt.txt","r")
print(fileobj.read())
```

```python
# 2. How would you read two characters from a file .

fileobj = open("Sample_txt.txt","r")
print(fileobj.read(2))
fileobj.close()
```

```python
# 3. Which command is used to read the entire content of the file as a string.

fileobj = open("Sample_txt.txt","r")
print(fileobj.read())
fileobj.close()
```

```python
# 4. How to read the next line of the file .

fileobj = open("Sample_txt.txt","r")
print(fileobj.readline())
print(fileobj.readline())
print(fileobj.readline())
print(fileobj.readline())
fileobj.close()
```
